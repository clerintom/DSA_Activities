### 1. The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.
 
 **Null Hypothesis, H0**: There is *no increase in sales* after stepping into digital marketing.

**Alternative Hypothesis, H1**: There is an *increase in sales* after stepping into digital marketing.

```python
from scipy import stats
```

```python
ttest,pval = stats.ttest_ind(df['Sales_After_digital_add(in $)'],df['Sales_before_digital_add(in $)'],alternative="greater")
print("p-value",f"{pval:.27f}")
if pval <0.05:
  print("We reject null hypothesis i.e. There is an increase in sales after stepping into digital marketing.")
else:
  print("We accept null hypothesis i.e. There is no increase in sales after stepping into digital marketing.")
```

**Inference::**
* Since the p-value < 0.05, which shows that we can reject the null hypothesis.

* So we can clearly say that the there is an increase in sales for the company after stepping into digital marketing.

### 2. The company needs to check whether there is any dependency between the features 'Region' and 'Manager'.

**Null Hypothesis, H0**: There is **no dependency** between the features 'Region' & 'Manager'

**Alternative Hypothesis, H1**: There is a **dependency** between the features 'Region' & 'Manager'

```python
'''pandas.crosstab: Compute a simple cross tabulation of two (or more) factors. 
By default computes a frequency table of the factors unless an array of values and an aggregation function are passed.'''
ctab = pd.crosstab(df.Region,df.Manager)
ctab
```

* #### **Since the region and manager features having categorical values, we can perform chi-square test.**

```python
from bioinfokit.analys import stat

# run chi-square test for independence
result=stat()

result.chisq(df=ctab)
# output
print(result.summary)

```

**Inference:**
* Since the p-value is > 0.05, we can accept the null hypothesis.
* So we can say that the features "Region" and "Manager" are independent to each other. There is no relation between the two features.

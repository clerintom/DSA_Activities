###   1. Reading the dataset 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("D:\Clerin\data scienc\cars_data.csv") 

df
```

###  2. Null values present in the dataset

```python
df.isnull().head()
```

```python
# sum  of missing values
df.isna().sum()
```

###  3. Countplot of Buyer Gender in the sale

```python
import warnings
warnings.filterwarnings("ignore")
sns.countplot(x="Buyer Gender",data= df,hue="Buyer Gender")
plt.legend(loc='upper center')
```

###  4.Top Five cars based on  sales price

```python
max=df.nlargest(5, ['Sale Price'])
print(max[["Make","Model","Nickname","Country","Color","Sale Price"]])
```

### 5. Least Five cars based on their Resell price

```python
min=df.nsmallest(5, ['Resell Price'])
print(min[["Make","Model","Nickname","Country","Color","Resell Price"]])
```

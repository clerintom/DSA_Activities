## READ DATA SET

```python
import pandas as pd 

df = pd.read_csv("D:\Clerin\data scienc\week4\diabetes_model.csv") ## as csv format
print(df)

```

### 1.Create a Count plot of Outcome

```python
import seaborn as sns
sns.countplot(x ='Outcome', data = df)
```

### 2.Effect of age on Outcome using Boxplot

```python
import matplotlib.pyplot as plt

sns.boxplot(x='Outcome',y='Age',data=df)
plt.title('Age on Outcome')
plt.show()
# show plot
plt.show()
```

### 3.Chances of getting diabetes, using the Diabetes Pedigree Function 

```python
sns.relplot(x='Outcome', y='DiabetesPedigreeFunction', data=df)
 
# giving title to the plot
plt.title('Graph')
 
# function to show plot
plt.show()
```
